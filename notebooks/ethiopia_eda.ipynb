```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../data/ethiopia.csv')
```

```python
dataset['country'] = 'Ethiopia'
dataset.head
```

```python
dataset['Date'] = pd.to_datetime(dataset['YEAR']*1000 + dataset['DOY'],format='%Y%j')
```

```python
dataset['Month'] = dataset['Date'].dt.month
dataset.head
```

```python
dataset.replace(-999,np.nan, inplace=True)
```

```python
dataset.duplicated().sum()
```

```python
dataset.describe() 
```

```python
duplicates = dataset.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
if duplicates > 0:
    dataset.drop_duplicates(inplace=True)
    print("Duplicates dropped.")
# Note: df.duplicated() checks entire rows; if duplicates exist, they span all columns.
```

```python
numeric_cols = dataset.select_dtypes(include=[np.number]).columns
dataset[numeric_cols].describe()
```

```python
missing = dataset.isna().sum()
missing_pct = (missing / len(dataset)) * 100
print(missing_pct)
high_missing = missing_pct[missing_pct > 5]
print(f"Columns with >5% missing: {list(high_missing.index)}")
# Interpretation: High missing values (e.g., >5%) may indicate data collection gaps, sensor failures, or regional unavailability, potentially biasing analysis if not handled (e.g., via imputation or exclusion).
```

```python
cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']
z_scores = (dataset[cols] - dataset[cols].mean()) / dataset[cols].std(ddof=0)
outlier_mask = np.abs(z_scores) > 3
outlier_count = outlier_mask.any(axis=1).sum()

print(f"Rows with |Z| > 3: {outlier_count}")
```

```python
outlier_rows = dataset[outlier_mask.any(axis=1)]
outlier_rows[['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']].head()
```

```python
weather_cols = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M', 'WS2M_MAX']
dataset[weather_cols] = dataset[weather_cols].ffill()
```

```python
threshold = int(0.3 * len(dataset.columns))
dataset = dataset.dropna(thresh=len(dataset.columns) - threshold)
```

```python
row_missing_pct = dataset.isna().mean(axis=1)
dataset = dataset[row_missing_pct <= 0.30]
```

```python
dataset.to_csv('../data/ethiopia_clean.csv', index=False)
```
